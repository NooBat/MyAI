```python
import numpy as np
import matplotlib.pyplot as plt

def get_tangent_line(x) :
    np_x = np.arange(x - 1, x + 1, 0.01)
    y = (2 * x + 2) * (np_x - x) + (x ** 2 + 2 * x + 1)
    return (np_x, y)

x = np.arange(-5, 3, 0.01)
y = x ** 2 + 2 * x + 1

xlab = r'$\theta_1$'
ylab = 'J(' + r'$\theta_1$' + ')'
plt.xlabel(xlab)
plt.ylabel(ylab)

plt.plot(x, y)

theta = -4.03993

#we choose learning rate 𝞪 = 0.05

learn_rate = 0.1

theta_lst = [theta]

while theta > -0.99999 or theta < -1.00001:
    theta = theta - learn_rate * (theta * 2 + 2)
    theta_lst.append(theta)

print(len(theta_lst))
np_theta_array = np.array(theta_lst)
np_y = np_theta_array ** 2 + np_theta_array * 2 + 1

i = 0
while i < len(np_y):
    plt.scatter(np_theta_array[i], np_y[i])
    line_x, line_y = get_tangent_line(np_theta_array[i])
    plt.plot(line_x, line_y)
    i += 1
    
plt.show()
```
